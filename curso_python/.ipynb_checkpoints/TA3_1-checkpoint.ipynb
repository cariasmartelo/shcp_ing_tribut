```python
columns = ['isr_neto', 'iva_neto']
ingresos[columns].head()
```

### Creamos ingresos totales

```python
ingresos['ingresos_totales'] = ingresos['isr_neto']\
        + ingresos['iva_neto'] + ingresos['ieps_neto']
```

```python
ingresos['ingresos_totales'] = \
    ingresos[['isr_neto', 'iva_neto', 'ieps_neto']].sum(axis=1)
```

```python
ingresos.head()
```

### Creamos cada variable como porcentaje del total

```python
ingresos['isr_perc'] = ingresos['isr_neto'] / ingresos['ingresos_totales']
ingresos['iva_perc'] = ingresos['iva_neto'] / ingresos['ingresos_totales']
ingresos['ieps_perc'] = ingresos['ieps_neto'] / ingresos['ingresos_totales']
```

```python
ingresos
```

### Usemos loc con condiciones

```python
ingresos.loc[ingresos['isr_neto'] > 100000, 'isr_neto']
```

```python
ingresos.loc[(ingresos[['isr_neto', 'iva_neto']] < 30000).all(1), 'isr_neto']
```

```python
condicion = (ingresos['isr_neto'] < 40000) & (ingresos['iva_neto'] < 30000)
ingresos.loc[condicion, 'isr_neto']
```
