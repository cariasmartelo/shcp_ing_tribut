```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import funcion_importar_csv
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
ingresos_netos = funcion_importar_csv.importar_ingresos()
```

```python
ingresos_netos.columns
```

Graficamos

```python
cols_to_plot = ['ingresos_sector_publico_neto_(mdp)_r', 'ingresos_gobierno_federal_neto_(mdp)_r', 
                'ingresos_tributarios_neto_(mdp)_r', 'ingresos_no_tributarios_neto_(mdp)_r',
                'isr_neto_(mdp)_r', 'iva_neto_(mdp)_r', 'importaciones_neto_(mdp)_r',
                'ieps_sin_gas_neto_(mdp)_r','ieps_gasolina_neto_(mdp)_r']
```

# Método directo

```python
ingresos_netos[cols_to_plot].plot()
```

# Método que permite hacer modificaciones

```python
fig, ax = plt.subplots()
ax.plot(ingresos_netos[cols_to_plot])
plt.show()
```

Añadimos legenda

```python
fig, ax = plt.subplots()
ax.plot(ingresos_netos[cols_to_plot])
ax.legend(cols_to_plot)
plt.show()
```

Incrementamos tamanño

```python
fig, ax = plt.subplots(figsize=(15, 10))
ax.plot(ingresos_netos[cols_to_plot])
ax.legend(cols_to_plot)
plt.show()
```

Creamos una legenda específica

```python
legend_dict = {'ingresos_sector_publico_neto_(mdp)_r': 'Ingresos sector público',
               'ingresos_gobierno_federal_neto_(mdp)_r': 'Ingresos Gobierno Federal', 
               'ingresos_tributarios_neto_(mdp)_r': 'Ingresos Tributarios',
               'ingresos_no_tributarios_neto_(mdp)_r': 'Ingresos no tributarios',
               'isr_neto_(mdp)_r': 'ISR',
               'iva_neto_(mdp)_r': 'IVA',
               'importaciones_neto_(mdp)_r': 'Impuesto a las importaciones',
               'ieps_sin_gas_neto_(mdp)_r': 'IEPS sin gasolina',
               'ieps_gasolina_neto_(mdp)_r': 'IEPS a las gasolinas'}
```

```python
[legend_dict[col] for col in cols_to_plot]
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
ax.plot(ingresos_netos[cols_to_plot])
legend = [legend_dict[col] for col in cols_to_plot]
ax.legend(legend)
plt.show()
```

Anñadimos grid y título

```python
fig, ax = plt.subplots(figsize=(15, 10))
ax.plot(ingresos_netos[cols_to_plot])
legend = [legend_dict[col] for col in cols_to_plot]
ax.legend(legend)
ax.set_title('Ingresos del Sector Público en México\n(MDP de 2019)')
ax.grid()
plt.show()
```

Incrementamos calidad

```python
fig, ax = plt.subplots(figsize=(15, 10), dpi=100)
ax.plot(ingresos_netos[cols_to_plot])
legend = [legend_dict[col] for col in cols_to_plot]
ax.legend(legend)
ax.set_title('Ingresos del Sector Público en México\n (MDP de 2019)')
ax.grid()
plt.show()
```

Cambiamos tipo de linea

```python
fig, ax = plt.subplots(figsize=(15, 10), dpi=100)
ax.plot(ingresos_netos[cols_to_plot], linestyle='--')
legend = [legend_dict[col] for col in cols_to_plot]
ax.legend(legend)
ax.set_title('Ingresos del Sector Público en México\n (MDP de 2019)')
ax.grid()
plt.show()
```

```python
linestyles_dict = {'ingresos_sector_publico_neto_(mdp)_r': '-',
               'ingresos_gobierno_federal_neto_(mdp)_r': '--', 
               'ingresos_tributarios_neto_(mdp)_r': ':',
               'ingresos_no_tributarios_neto_(mdp)_r': '-.',
               'isr_neto_(mdp)_r': '-',
               'iva_neto_(mdp)_r': '-',
               'importaciones_neto_(mdp)_r': '--',
               'ieps_sin_gas_neto_(mdp)_r': ':',
               'ieps_gasolina_neto_(mdp)_r': ':'}
colors_dict = {'ingresos_sector_publico_neto_(mdp)_r': 'k',
               'ingresos_gobierno_federal_neto_(mdp)_r': 'b', 
               'ingresos_tributarios_neto_(mdp)_r': 'r',
               'ingresos_no_tributarios_neto_(mdp)_r': 'y',
               'isr_neto_(mdp)_r': 'c',
               'iva_neto_(mdp)_r': 'm',
               'importaciones_neto_(mdp)_r': 'y',
               'ieps_sin_gas_neto_(mdp)_r': '#333f50',
               'ieps_gasolina_neto_(mdp)_r': '#691A30'}
```

```python
fig, ax = plt.subplots(figsize=(15, 10), dpi=100)
for col in cols_to_plot:
    ax.plot(ingresos_netos[col], linestyle=linestyles_dict[col], color = colors_dict[col])
legend = [legend_dict[col] for col in cols_to_plot]
ax.legend(legend)
ax.set_title('Ingresos del Sector Público en México\n (MDP de 2019)')
ax.grid()
plt.show()
```

# Grafica PIE

```python
cols_tributarios = ['isr_neto_(mdp)_r', 'iva_neto_(mdp)_r', 'importaciones_neto_(mdp)_r', 'ieps_sin_gas_neto_(mdp)_r',
                    'ieps_gasolina_neto_(mdp)_r']
```

```python
ingresos_netos_y = ingresos_netos.resample('YS').sum()
```

```python
ingresos_2018 = ingresos_netos_y.loc['2018-01-01']
```

```python
ingresos_2018[cols_tributarios]
```

```python
fig, ax = plt.subplots(figsize=(15, 10), dpi=100)
ax.pie(ingresos_2018[cols_tributarios])
legend = [legend_dict[col] for col in cols_tributarios]
ax.legend(legend)
ax.set_title('Ingresos del Sector Público en México\n (MDP de 2019)')
plt.show()
```
