```python
REVENUE_EXCEL = '../inputs/bases_iniciales/Análisis_y_base_ingresos_stata_mens_v190508_v1730.xlsx'
params={
    
}
```

```python
revenue_df = pd.read_excel(REVENUE_EXCEL)
```

```python
revenue_df.rename(columns={'anio': 'year', 'mes':'month', 'ing_tot_brut': 'ing_tot_bruto',
                           'ing_trib_brut': 'ing_trib_bruto', 'ing_neto': 'ing_tot_neto',
                           'iva_regu': 'iva_reg'},
                  inplace=True)
revenue_df['day'] = 1
revenue_df['date'] = pd.to_datetime(revenue_df[['year', 'month', 'day']])
revenue_df.set_index('date', inplace=True)
```

```python
from_pesos_to_thousand_million = ['ing_tot_bruto', 'ing_tot_neto', 'ing_trib_bruto', 'ing_trib_neto',\
                                  'ing_no_trib_bruto', 'ing_no_trib_neto', 'iva_bruto', 'iva_neto',\
                                  'iva_dev', 'iva_comp', 'iva_reg', 'isr_bruto', 'isr_dev', 'isr_comp',\
                                  'isr_reg', 'isr_neto', 'ieps_bruto', 'ieps_neto', 'ieps_dev',\
                                  'ieps_comp', 'ieps_reg']
revenue_df[from_pesos_to_thousand_million] = revenue_df[from_pesos_to_thousand_million]/1000000000
```

```python
to_real = ['ing_tot_bruto', 'ing_tot_neto', 'ing_trib_bruto', 'ing_trib_neto', 'ing_no_trib_bruto',\
           'ing_no_trib_neto', 'iva_bruto', 'iva_neto', 'iva_dev', 'iva_comp', 'iva_reg', 'isr_bruto',\
           'isr_dev', 'isr_comp', 'isr_reg', 'isr_neto', 'ieps_bruto', 'ieps_neto', 'ieps_dev', 'ieps_comp',\
           'ieps_reg']
for col in to_real:
    revenue_df[col+'_r'] = (revenue_df[col] / revenue_df['inpc'])
```

```python
revenue_df.columns
```

```python
federal_income = ['ing_tot_', 'ing_trib_', 'ing_no_trib_']
income_taxes = ['ing_trib_', 'iva_', 'isr_', 'ieps_']
tax_decomposition = ['bruto', 'neto', 'dev', 'comp', 'reg']
cols = {
    'gross_income_cols': [col + 'bruto' for col in federal_income],
    'net_income_cols': [col + 'neto' for col in federal_income],
    'gross_tax_income_cols': [col + 'bruto' for col in income_taxes],
    'net_tax_income_cols': [col + 'neto' for col in income_taxes],
    'iva_cols': ['iva_' + col for col in tax_decomposition],
    'isr_cols': ['isr_' + col for col in tax_decomposition],
    'ieps_cols': ['ieps_' + col for col in tax_decomposition]
}
real_cols = {}
for key, columns in cols.items():
    real_cols[key + '_r'] = [col + '_r' for col in columns]
    
for key, columns in real_cols.items():
    cols[key] = columns
```

```python
for k, columns in cols.items():
    print(set(columns).issubset(set(revenue_df.columns)))
```

## Nominal Levels
Nominal analysis of Public revenue and tax income

```python
years = mdates.YearLocator()   # every year
quarters = mdates.MonthLocator(bymonth = range(1, 13, 3))  # every month
date_fmt = mdates.DateFormatter('%Y-%m')
fig, ax = plt.subplots(nrows=4, figsize=(15,24), sharex=False)
plot_names = ['Nominal Gross Income', 'Nominal Gross Tax Income', 'Nominal Net Tax Income']
i = 0
for k, cols_to_plot in cols.items():
    if k in ['gross_income_cols', 'gross_tax_income_cols', 'net_income_cols', 'net_tax_income_cols']:
        ax[i].plot(revenue_df[cols_to_plot])
        ax[i].legend(cols_to_plot)
        ax[i].xaxis.set_major_locator(quarters)
        ax[i].xaxis.set_major_formatter(date_fmt)
        ax[i].grid()
        ax[i].set_title(k)
        ax[i].set_ylabel('MMDP')
        for tick in ax[i].get_xticklabels():
            tick.set_rotation(40)
        i += 1
plt.show()
```

## Nominal Percentual
Nominal analysis of Public revenue and tax income in percentual terms

```python
years = mdates.YearLocator()   # every year
quarters = mdates.MonthLocator(bymonth = range(1, 13, 3))  # every month
date_fmt = mdates.DateFormatter('%Y-%m')
fig, ax = plt.subplots(nrows=4, figsize=(15,24), sharex=False)
plot_names = ['Perc of Gross Income', 'Perc of Net Income', 'Perc of Gross Tax Income', 'Perc of Net Tax Income']
i = 0
for k, cols_to_plot in cols.items():
    if k in ['gross_income_cols', 'net_income_cols', 'gross_tax_income_cols', 'net_tax_income_cols']:
        to_plot = revenue_df[cols_to_plot[1:]].div(revenue_df[cols_to_plot[0]], axis=0)
        ax[i].plot(to_plot)
        ax[i].legend(to_plot.add_suffix('_perc').columns)
        ax[i].xaxis.set_major_locator(quarters)
        ax[i].xaxis.set_major_formatter(date_fmt)
        ax[i].grid()
        ax[i].set_title(plot_names[i])
        for tick in ax[i].get_xticklabels():
            tick.set_rotation(40)
        i += 1
plt.show()
```

## Real IVA and ISR
Real analysis of IVA and ISR

```python
years = mdates.YearLocator()   # every year
quarters = mdates.MonthLocator(bymonth = range(1, 13, 3))  # every month
date_fmt = mdates.DateFormatter('%Y-%m')
fig, ax = plt.subplots(nrows=2, figsize=(15,12), sharex=False)
i = 0
for k, cols_to_plot in cols.items():
    if k in ['gross_tax_income_cols_r', 'net_tax_income_cols_r']:
        ax[i].plot(revenue_df[cols_to_plot])
        ax[i].legend(cols_to_plot)
        ax[i].xaxis.set_major_locator(quarters)
        ax[i].xaxis.set_major_formatter(date_fmt)
        ax[i].grid()
        ax[i].set_title(k)
        ax[i].set_ylabel('MMDP2019')
        for tick in ax[i].get_xticklabels():
            tick.set_rotation(40)
        i += 1
plt.show()
```

```python
years = mdates.YearLocator()   # every year
quarters = mdates.MonthLocator(bymonth = range(1, 13, 3))  # every month
date_fmt = mdates.DateFormatter('%Y-%m')
fig, ax = plt.subplots(figsize=(15,12), sharex=False)
i = 0
ls = '-'
labels = []
for k, cols_to_plot in cols.items():
    if k in ['gross_tax_income_cols_r', 'net_tax_income_cols_r']:
        plt.gca().set_prop_cycle(None)
        ax.plot(revenue_df[cols_to_plot], ls=ls)
        ls = '--'
        labels += cols_to_plot
ax.legend(labels)
ax.xaxis.set_major_locator(quarters)
ax.xaxis.set_major_formatter(date_fmt)
ax.grid()
ax.set_title(k)
ax.set_ylabel('MMDP2019')
ax.set_yticks(np.arange(0, 450, 50))
for tick in ax.get_xticklabels():
    tick.set_rotation(40)
plt.show()
```
