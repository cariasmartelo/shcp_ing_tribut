## Decomposition

```python
rcParams['figure.figsize'] = 18, 8
decomposition = sm.tsa.seasonal_decompose(fiscal_income_df[params['outcome_col']], model='additive')
fig = decomposition.plot()
plt.show()
type(decomposition)
```

Se ve tendencia positiva en ingresos tributarios y estacionalidad. 

# Estacionalidad

```python
descriptive.plot_series(decomposition.seasonal, title='Estacionalidad de Ingresos Tributarios Federales', 
                        subtitle='(MMDP 2018)', figsize=(15, 8), legend=False)
```

Se ve tendencia positiva de ingresos fiscales, as como fuerte componente estacional,

## Testing for stationarity and transforming

```python
descriptive.test_stationarity(fiscal_income_df, params['outcome_col'])
```

```python
transforms = ['log', 'diff', 'log_diff']
for t in transforms:
    print('\n \n TRANSFORMATION {} \n'.format(t))
    new_serie_name = params['outcome_col'] + '_' + t
    fiscal_income_df[new_serie_name] = descriptive.transformation(fiscal_income_df, params['outcome_col'], t)
    descriptive.test_stationarity(fiscal_income_df, new_serie_name)
```

## Stationarity results
- level: Not stationary
- log: Not stationary
- diff: Stationary
- log_diff: Stationary

Vamos a usar log diff para el anlisis

```python
objective_ts = fiscal_income_df[params['outcome_col'] + '_' + 'log_diff']
objective_ts = objective_ts[objective_ts.notna()]
```

# ARMA

```python
descriptive.plot_acf_pacf(objective_ts, 20)
```

```python
arma_order_select_ic(objective_ts, 3, 3, ic=['aic', 'bic'], trend = 'c')
```

```python
arma_model = ARMA(objective_ts, (2, 3))
results_arma = arma_model.fit()
fitted_arma = results_arma.fittedvalues
plt.plot(objective_ts, label = 'True')
plt.plot(fitted_arma, color = 'r', label='Fitted')
plt.show()
rss_arma = results_arma.resid ** 2
rmse_arma = np.sqrt(rss.mean())
mae_arma = abs(results_arma.resid).mean()
```

```python
arima_model = ARIMA(objective_ts, (2,0,3))
results_arima = arima_model.fit()
fitted_arima = results_arma.fittedvalues
prediction = results_arima.predict(start='2019-04-01', end='2022-04-01')
plt.plot(objective_ts, label = 'True')
# plt.plot(fitted_arima, color = 'r', label='Fitted')
plt.plot(prediction, color='k')
plt.show()
rss_arima = results_arima.resid ** 2
rmse_arima = np.sqrt(rss.mean())
mae_arima = abs(results_arma.resid).mean()
```
