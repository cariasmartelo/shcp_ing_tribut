# Holt Winters

```python
taxes_to_predict = [col for col in ingresos_netos.columns if col.endswith('_r')]
```

```python
models = {}
ingresos_netos_pred = ingresos_netos.loc[:, taxes_to_predict].copy()
for tax in taxes_to_predict:
    ts = ingresos_netos[tax]
    ts = ts[ts.notna()]
    end_of_series = ts.index.max()
    # COnstruimos modelo
    print(tax)
    print(ts.shape)
    model = ExponentialSmoothing(ts, seasonal='add',trend='add', seasonal_periods=12)
    model = model.fit()
    models[tax + '_HW'] = model

    # Construimos las predicciones
    start_prediction = end_of_series + relativedelta(months=1)
    pred = model.predict(start=start_prediction, end='2020-12-01')
    ts_pred = pd.concat([ts, pred], sort=True)
    ingresos_netos_pred = ingresos_netos_pred.merge(ts_pred.rename(tax + '_HW'), left_index=True, right_index=True,
                                                    how='outer') 
```

```python
exog = pd.DataFrame(index=ingresos_netos_dif.index)
exog = exog.reindex(pd.date_range(exog.index.min(), '2020-12-01', freq='MS'))
exog['reforma14'] = np.where(exog.index.year == 2015, 1, 0)
exog['rec97'] = np.where(exog.index.year == 1997, 1, 0)
exog['rec10'] = np.where(exog.index.year == 2010, 1, 0)
exog['cons'] = 1
```
