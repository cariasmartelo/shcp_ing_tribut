```python
import pandas as pd
import numpy as np
import download
import descriptive
```

```python
indic_mens_consumo = download.load_inegi_indic('indic_mens_consumo')
emec_menor_total = download.load_inegi_indic('emec_menor_total')
emec_menor_aba_ali_beb_tab = download.load_inegi_indic('emec_menor_aba_ali_beb_tab')
emec_menor_aba_ali = download.load_inegi_indic('emec_menor_aba_ali')
emec_menor_hie_tab = download.load_inegi_indic('emec_menor_hie_tab')
emec_menor_antad = download.load_inegi_indic('emec_menor_antad')
emer_menor_text_vest_calz = download.load_inegi_indic('emer_menor_text_vest_calz')
emec_menor_pape_espar_otros =download.load_inegi_indic('emec_menor_pape_espar_otros')
emec_menor_domesticos = download.load_inegi_indic('emec_menor_domesticos')
emec_menor_vehic = download.load_inegi_indic('emec_menor_vehic')
importaciones =download.load_balanza_comercial('importaciones')
exportaciones =download.load_balanza_comercial('exportaciones')
```

```python
ingresos_totales = download.load_ingresos_fiscales('../inputs/ingresos_tributarios_desglosados_updated.xlsx')
indic_cons_anual = indic_mens_consumo.resample('YS').mean()
emec = pd.concat([emec_menor_total, emec_menor_aba_ali_beb_tab, emec_menor_aba_ali, emec_menor_hie_tab,
                  emec_menor_antad, emer_menor_text_vest_calz, emec_menor_pape_espar_otros,
                  emec_menor_domesticos, emec_menor_vehic], axis=1)
emec_anual = emec.resample('YS').mean()
ingresos_anuales = ingresos_totales.resample('YS').sum()
importaciones_anuales = importaciones['importaciones_r'].resample('YS').sum()
```

```python
emecs = ['emec_menor_total', 'emec_menor_aba_ali_beb_tab', 'emec_menor_aba_ali', 'emec_menor_hie_tab',
         'emec_menor_antad', 'emer_menor_text_vest_calz']
ieps = ['ieps_neto_(mdp)_r', 'ieps_sin_gas_neto_(mdp)_r', 'ieps_hidrocarburos_neto_(mdp)_r',
        'ieps_otros_petroliferos_neto_(mdp)_r', 'ieps_gasolina_neto_(mdp)_r', 'ieps_tabaco_neto_(mdp)_r',
        'ieps_ba_neto_(mdp)_r', 'ieps_cerveza_neto_(mdp)_r', 'ieps_telecom_neto_(mdp)_r',
        'ieps_otras_retenciones_neto_(mdp)_r', 'ieps_jys_neto_(mdp)_r', 'ieps_be_neto_(mdp)_r',
        'ieps_bsaborizadas_neto_(mdp)_r', 'ieps_aadc_neto_(mdp)_r', 'ieps_plaguicidas_neto_(mdp)_r',
        'ieps_combustibles_fosiles_neto_(mdp)_r', 'ieps_bruto_(mdp)_r']
ieps_relevantes = ['ieps_sin_gas_neto_(mdp)_r', 'ieps_tabaco_neto_(mdp)_r',
                   'ieps_ba_neto_(mdp)_r', 'ieps_cerveza_neto_(mdp)_r', 'ieps_telecom_neto_(mdp)_r',
                   'ieps_jys_neto_(mdp)_r', 'ieps_bsaborizadas_neto_(mdp)_r', 'ieps_aadc_neto_(mdp)_r']
ieps_sin_gasolina = ['ieps_sin_gas_neto_(mdp)_r', 'ieps_tabaco_neto_(mdp)_r',
                     'ieps_ba_neto_(mdp)_r', 'ieps_cerveza_neto_(mdp)_r', 'ieps_telecom_neto_(mdp)_r',
                     'ieps_jys_neto_(mdp)_r', 'ieps_bsaborizadas_neto_(mdp)_r', 'ieps_aadc_neto_(mdp)_r',
                     'ieps_plaguicidas_neto_(mdp)_r', 'ieps_otras_retenciones_neto_(mdp)_r']
```

```python
descriptive.plot_series(ingresos_totales, ['ieps_tabaco_neto_(mdp)_r'])
```

```python
df_anual = pd.concat([ingresos_anuales, importaciones_anuales, emec_anual, indic_cons_anual], axis=1)
df_anual = df_anual.loc[:'2018-01-01']
```

```python
descriptive.plot_series(df = df_anual, 
                        cols=ieps_sin_gasolina,
                       title='Recaudaci√≥n anual por IEPS',
                       subtitle='(MDP de 2019)')
```

```python
descriptive.plot_series(df = df_anual, 
                        cols=emecs,
                        title='Indicador de la Encuesta Mensual sobre Empresas Comerciantes',
                        subtitle='(2013 = 100)',
                        min_date='2012-01-01')
```

```python
# Creamos indice
index_2014 = pd.date_range('2014-01-01', '2014-12-01', freq='MS')
df_anual_index = df_anual.apply(lambda x: x/x[index_2014].mean() * 100)
```

```python
descriptive.plot_series(df = df_anual_index, 
                        cols=ieps_relevantes,
                       title='Recaudaci√≥n anual por IEPS',
                       subtitle='(MDP de 2019)',
                       min_date='2014-01-01')
```
