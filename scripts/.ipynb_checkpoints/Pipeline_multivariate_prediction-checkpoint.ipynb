```python
# Vamos cuales son las variables que tenemos que hacer nowcast
vars_to_predict = df_netos.columns[df_netos.tail().isna().any()]
vars_to_predict
```

```python
# Hacemos nowcast con ARIMA y todas las demás variables disponibles como exógenas
models_nowcast = {}
# Predecimos variables usando ARIMA
for var in ['asegurados_imss', 'imai', 'igae']:
    # Obtenemos variables que usaremos como exógenas
    vars_to_use_as_covars = df_netos.columns[df_netos.tail().notna().all()]
    # Creamos df de variables exógenas
    exog = df_netos[vars_to_use_as_covars].copy()
    # Creamos variable de mes
    exog['mes'] = exog.index.map(lambda x: x.month)
    # Creamos dummies de meses
    month_dummies = pd.get_dummies(exog['mes'])
    month_dummies = month_dummies.add_prefix('mes_')
    # Añadimos las dummies de meses al df de exogenas
    exog = pd.concat([exog, month_dummies], axis=1)
    # Creamos dummie para 1995
    exog['crisis_2009'] = np.where(exog.index.year == 2009, 1, 0)
    # Eliminamos columnas mes
    exog.drop('mes', axis=1, inplace=True)
    # Creamos time series que usaremos para predecir
    ts = df_netos[var].copy()
    # Eliminamos los valores NA
    ts = ts[ts.notna()]
    # Creamos diferencia
    ts_d = ts.diff(12)
    # Elimiamos NA's de las diferencia
    ts_d = ts_d[ts_d.notna()]
    # Obtenemose la fecha del ultimo valor no NA
    end_of_series = ts_d.index.max()
    # Uamos ese valor para filtrar el DF de variables exógenas.
    exog_train = exog.loc[exog.index <= end_of_series].copy()
    # Quitamos columnas con NAS del df exogeno
    exog_train = exog_train.loc[:, exog_train.notna().all()]
    # Obtenemos el exog_test a partir de la fecha y con las columnas ya filtradas
    exog_test = exog.loc[exog.index > end_of_series, exog_train.columns].copy()
    # Nos aseguramos que tengan el mismo indice exog y ts_d
    exog_train_model = exog_train.loc[ts_d.index]
    # Construimos modelo
    model = sm.tsa.statespace.SARIMAX(
            ts_d, order=(3, 0, 3), seasonal_order=(1, 0, 1, 12),
            exog=exog_train_model, enforce_stationarity=False, enforce_invertibility=False)
    model = model.fit()
    models_nowcast[var + '_SARIMA'] = model

    # Construimos las predicciones
    # Creamos variables start prediction
    start_prediction = end_of_series + relativedelta(months=1)
    # Creamos periodo de prediccion
    prediction_period = pd.date_range(start_prediction, df_netos.index.max(), freq='MS')
    # Creamos prediccion
    pred = model.predict(start=start_prediction, end=df_netos.index.max(), exog=exog_test)
    # Concatenamos prediccion de diferencias con observado
    ts_pred = pd.concat([ts, pred], sort=True)
    # Convertimos la prediccion a niveles sumando el valor de la varaible 12 meses antes. Hacemos loop.
    # Cuando haces loop sobre una seriel se especifica iteritems(), lo cual da un tuple de
    # (indice, valor). Los abstraemos como d, val.
    for d, val in ts_pred[prediction_period].iteritems():
        ts_pred[d] = val + ts_pred.shift(12)[d]
    # Reemplazamos la variable en df netos con la predicción
    df_netos[var] = ts_pred
    print('Nowcasted {}'.format(var))
```

# Predecimos variables de Estados Unidos hasta 2020 usando un modelo VAR

```python
us_variables = ['tbill_3meses_mensual', 'cons_price_index_us', 'ind_prod_ind_us']
```

```python
# Analizamos estacionariedad
for var in us_variables:
    ts = df_netos[var]
    descriptive.test_stationarity(ts)
```

```python
# Analizamos estacionariedad. Transformamos a primera diferencia
for var in us_variables:
    ts = df_netos[var]
    ts_log_diff = descriptive.transformation(ts, 'log_diff')
    ts_log_diff = ts_log_diff.rename('dif_log_{}'.format(var))
    descriptive.test_stationarity(ts_log_diff)
```

```python
# Creamos DF para modelo VAR
var_us = df_netos[us_variables].apply(descriptive.transformation, args=('log_diff', 1))
var_us = var_us.loc[var_us.notna().all(1)]
```

```python
# Hacemos prueba de cointegración
johansen = select_coint_rank(var_us, 0, 12, method='trace')
johansen.rank
```

```python
# Creamos variables exogenas
var_us_exog = s_exog = pd.DataFrame(index = pd.date_range(var_us.index.min(), '2020-12-01', freq='MS'))
# # Creamos variable de mes
# var_us_exog['mes'] = var_us_exog.index.map(lambda x: x.month)
# # Creamos dummies de meses
# month_dummies = pd.get_dummies(var_us_exog['mes'])
# month_dummies = month_dummies.add_prefix('mes_')
# # Añadimos las dummies de meses sl df de exogenas
# var_us_exog = pd.concat([var_us_exog, month_dummies], axis=1)
# Creamos dummie para 2009
var_us_exog['crisis_2009'] = np.where(var_us_exog.index.year == 2009, 1, 0)
# Eliminamos columnas mes
# var_us_exog.drop('mes', axis=1, inplace=True)
# # Dividimos exog en exog_train y exog_test
var_us_exog_train = var_us_exog.loc[var_us.index].copy()
# # # Eliminmos NA
var_us_exog_train = var_us_exog_train.loc[var_us_exog_train.notna().all(1)]
var_us_exog_test = var_us_exog.loc[var_us.index.max() + relativedelta(months=1):].copy()
```

```python
var_us = var_us.loc[var_us_exog_train.index]
```

```python
# Corremos modelo y obtenemos predicción
model_us = VAR(var_us, exog=var_us_exog_train)
results_us = model_us.fit(maxlags=12, ic='aic')
lags_used_us = results_us.k_ar
us_forecast  = results_us.forecast(y=var_us.values[-lags_used_us:],steps=18, exog_future=var_us_exog_test)
```

```python
# Creamos un DF con las predicciones
prediction_us = pd.DataFrame()
for i, var in enumerate(us_variables):
    var_d_pred = us_forecast[:,i]
    var_d_pred = pd.Series(var_d_pred, index=pd.date_range('2019-07-01', '2020-12-01', freq='MS'))
    var_d_pred = pd.concat([var_us[var], var_d_pred], sort=True)
    initial_date = var_d_pred.index.min() - relativedelta(months=1)
    initial_value = df_netos.loc[initial_date, var]
    var_pred = descriptive.revert_transformation(var_d_pred, 'log_diff', initial_value, initial_date)
    prediction_us[var] = var_pred
```

```python
ml_exog = pd.concat([prediction_mex, prediction_us], axis=1)
ml_exog_pct_change = ml_exog.pct_change()
ml_exog_pct_change = ml_exog_pct_change.add_prefix('pct_change_')
ml_exog = pd.concat([ml_exog, ml_exog_pct_change], axis=1)
ml_exog['mes'] = ml_exog.index.map(lambda x: x.month)
# Creamos dummies de meses
month_dummies = pd.get_dummies(ml_exog['mes'])
month_dummies = month_dummies.add_prefix('mes_')
# Añadimos las dummies de meses sl df de exogenas
ml_exog = pd.concat([ml_exog, month_dummies], axis=1)
lags_ml_exog = pd.DataFrame(index=ml_exog.index)
for lag in range(1, 13):
    lag_ml_exog = ml_exog.shift(lag)
    lag_ml_exog = lag_ml_exog.add_prefix('lag_')
    lag_ml_exog = lag_ml_exog.add_suffix('_{}'.format(lag))
    lags_ml_exog = pd.concat([lags_ml_exog, lag_ml_exog], axis=1)
ml_exog = pd.concat([ml_exog, lags_ml_exog], axis=1)
ml_exog = pd.merge(ml_exog, semana_santa, left_index=True, right_index=True)
```

```python
# Creamos ingresos a predecir que no tocaremos
ingresos_a_predecir = df_netos[ingresos_fiscales_vars].copy()
# Creamos copia que sí tocaremos
ingresos_a_predecir_predicted = ingresos_a_predecir.copy()
# Creamos diferencia de logaritmos de igresos a predecir.
d_ingresos_a_predecir = ingresos_a_predecir.apply(descriptive.transformation, args=('log_diff', 1))
```

```python
prediction_date = pd.to_datetime('2019-07-01')
while prediction_date <= pd.to_datetime('2020-12-01'):
    features_test_start_date = prediction_date - relativedelta(months=1)
    # Creamos array endogeno
    ml_endog = pd.concat([ingresos_a_predecir_predicted, d_ingresos_a_predecir.add_prefix('log_diff_')], axis=1)
    # Añadimos lags
    lags_ml_endog = pd.DataFrame(index=ml_endog.index)
    for lag in range(1, 13):
        lag_ml_endog = ml_endog.shift(lag).add_prefix('lag_').add_suffix('_{}'.format(lag))
        lag_ml_endog = pd.concat([lags_ml_exog, lag_ml_endog], axis=1)
    ml_endog = pd.concat([ml_endog, lags_ml_endog], axis=1)
    ml_endog = ml_endog.loc['2003-01-01':]
    # Creamos variable a predecir
    outcome_var = d_ingresos_a_predecir.shift(-1).copy()
    # Creamos variable de fecha para hacer el melt
    outcome_var['fecha'] = outcome_var.index
    # Hacemos el melt para tener una variable y no cuatro por fecha
    outcome_var = pd.melt(outcome_var, id_vars = ['fecha']).sort_values(['fecha', 'variable'])
    # Regresamos la fecha al índice
    outcome_var.index = outcome_var['fecha']
    outcome_var.drop('fecha', axis=1, inplace=True)
    # Creamos dummies de las variables a predecir
    dummies_taxes_to_predict = pd.get_dummies(outcome_var['variable'], prefix='to_predict')
    # Obtenemos los nombres de las summies que hicimos
    dummies_names = [col for col in dummies_taxes_to_predict]
    # Concatenamos las dummies con la variable outcome
    outcome_var = pd.concat([outcome_var, dummies_taxes_to_predict], axis=1)
    # Ordenamos las variables
    outcome_var = outcome_var[dummies_names + ['variable', 'value']]
    # contarenamos el df endogeno con el df de outcome.
    ml_endog_outcome = ml_endog.merge(outcome_var, left_index=True, right_index=True, how='left')
    # Concatenamos el df exógeno con el endógeno y el outcome.
    X_Y = pd.merge(ml_exog, ml_endog_outcome, left_index=True, right_index=True)
    # Obtenemos train y test
    train = X_Y.loc[X_Y.index < features_test_start_date]
    test = X_Y.loc[X_Y.index >= features_test_start_date]
    # Filtramos train y obtenemos X y Y usando el orden de las columnas
    train = train.loc[train.notna().all(1)]
    X_train = train.iloc[:,:-2]
    X_test = test.iloc[:,:-2]
    Y_train = train.iloc[:,-1]
    Y_test = test.iloc[:,-1]
    predicted_labels = test.iloc[:,-2]
    # Construimos la regresión
    regr = GradientBoostingRegressor(n_estimators=100, random_state=1234)
    regr = regr.fit(X_train, Y_train)
    prediction = regr.predict(X_test)
    # Hacemos diccionario y lo usamos para crear DF que concatenaremos
    dictionary_predictions = {key: prediction[i] for i, key in enumerate(predicted_labels.values)}
    predictions_to_append = pd.DataFrame(dictionary_predictions, index = [prediction_date])
    # Hacemos append
    d_ingresos_a_predecir = d_ingresos_a_predecir.append(predictions_to_append, sort=True)
    # Obtenemos valores iniciales para revertir
    start_transformed = d_ingresos_a_predecir.loc[d_ingresos_a_predecir.notna().all(1)].index.min()
    initial_date = start_transformed - relativedelta(months=1)
    initial_values = ingresos_a_predecir_predicted.loc[initial_date]
    # Construios de nuevo ingresos a predecir predicted
    ingresos_a_predecir_predicted = d_ingresos_a_predecir.apply(lambda x: descriptive.revert_transformation(
                                        x, 'log_diff', initial_values[x.name], initial_date))
    # Anñadimos 1 mes
    prediction_date = prediction_date + relativedelta(months=1)
```
